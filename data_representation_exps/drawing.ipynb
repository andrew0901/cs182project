```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
oh_mc_test = np.genfromtxt('oh_mc_test.csv', delimiter=',')
oh_mr_test = np.genfromtxt('oh_mr_test.csv', delimiter=',')
```

```python
oh_rs_train = np.genfromtxt('oh_rs_train.csv', delimiter=',')
oh_rs_test = np.genfromtxt('oh_rs_test.csv', delimiter=',')
```

```python
ohr_mc_test = np.genfromtxt('ohr_mc_test.csv', delimiter=',')
ohr_mr_test = np.genfromtxt('ohr_mr_test.csv', delimiter=',')
```

```python
ohr_rs_train = np.genfromtxt('ohr_rs_train.csv', delimiter=',')
ohr_rs_test = np.genfromtxt('ohr_rs_test.csv', delimiter=',')
```

```python
sg_mc_test = np.genfromtxt('sg_mc_test.csv', delimiter=',')
sg_mr_test = np.genfromtxt('sg_mr_test.csv', delimiter=',')
sg_rs_train = np.genfromtxt('sg_rs_train.csv', delimiter=',')
sg_rs_test = np.genfromtxt('sg_rs_test.csv', delimiter=',')
```

```python
sge_mc_test = np.genfromtxt('sge_mc_test.csv', delimiter=',')
sge_mr_test = np.genfromtxt('sge_mr_test.csv', delimiter=',')
sge_rs_train = np.genfromtxt('sge_rs_train.csv', delimiter=',')
sge_rs_test = np.genfromtxt('sge_rs_test.csv', delimiter=',')
```

```python
plt.figure()
plt.title("Loss vs Epochs For Random Split Train/Validation Set")
plt.plot(oh_rs_train, label="train one-hot everything", linewidth=2, color = "tab:blue")
plt.plot(oh_rs_test, label = "valn one-hot everything", linewidth=2, color = "tab:blue", linestyle = "dashed")
plt.plot(ohr_rs_train, label="train one-hot reactant", linewidth=2, color = "tab:orange")
plt.plot(ohr_rs_test, label = "val one-hot reactant", linewidth=2, color = "tab:orange", linestyle = "dashed")
plt.plot(sg_rs_train, label="train all graph", linewidth=2, color = "tab:green")
plt.plot(sg_rs_test, label = "val all graph", linewidth=2, color = "tab:green", linestyle = "dashed")
plt.plot(sge_rs_train, label="train all graph with edge", linewidth=2, color = "tab:purple")
plt.plot(sge_rs_test, label = "val all graph with edge", linewidth=2, color = "tab:purple", linestyle = "dashed")
plt.ylim([0.44, 0.55])
plt.legend(loc='center right',bbox_to_anchor=(1.5, 0.5))
plt.savefig('rs_data_representation.png')
plt.show()
```

```python
plt.figure()
plt.title("Validation Loss vs Epochs For Masked Catalyst")
plt.plot(oh_mc_test, label="One-hot everything", linewidth=2)
plt.plot(ohr_mc_test, label = "One-hot reactant", linewidth=2)
plt.plot(sg_mc_test, label="All graph", linewidth=2)
plt.plot(sge_mc_test, label = "All graph with edge")
#plt.ylim([0.44, 0.55])
plt.legend()
plt.savefig('mc_data_representation.png')
plt.show()
```

```python
plt.figure()
plt.title("Validation Loss vs Epochs For Masked Reactant")
plt.plot(oh_mr_test, label="One-hot everything", linewidth=2)
plt.plot(ohr_mr_test, label = "One-hot reactant", linewidth=2)
plt.plot(sg_mr_test, label="All graph", linewidth=2)
plt.plot(sge_mr_test, label = "All graph with edge")
#plt.ylim([0.44, 0.55])
plt.legend()
plt.savefig('mr_data_representation.jpg')
plt.show()
```
